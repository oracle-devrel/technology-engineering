# Creating and running jobs

```python
from ads.jobs import Job, DataScienceJob, PythonRuntime
from ads import set_auth
set_auth("resource_principal")
```

```python
job = (
    Job(name="anomaly_detection_scoring v.3")
    .with_infrastructure(
        DataScienceJob()
        .with_log_group_id("<group_id_ocid>")
        .with_log_id("<log_id_ocid>")
        .with_shape_name("VM.Standard.E4.Flex")
        .with_shape_config_details(memory_in_gbs=4, ocpus=1)
        .with_block_storage_size(50) # minimus is 50
    )
    .with_runtime(
        PythonRuntime()
        .with_service_conda("generalml_p311_cpu_x86_64_v1")
        .with_source("/home/datascience/code/anomaly_detection/fraud_detection/src/batch_inference.py")
        .with_environment_variable(FILE_NAME="batch-inference/day1_df.csv") 
    )
)
job.create()
```

```python
job_run = job.run(
    name="day1",
    env_var={'FILE_NAME': 'batch-inference/day1.csv'} # in every run we can change these values.
)
job_run.watch()
```

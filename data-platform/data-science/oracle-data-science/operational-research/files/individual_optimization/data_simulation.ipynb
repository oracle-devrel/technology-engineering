```python
import numpy as np
import pandas as pd
```

```python
def simulate_training_data(size):
    age = np.random.uniform(20, 70, size)
    risk = np.random.uniform(0, 1, size)
    price = np.random.uniform(2, 4,size)

    beta_0 = 3.0
    beta_price = -1.5
    beta_age = 0.05
    beta_risk = -1.0
    
    logit = (
        beta_0
        + beta_price * price
        + beta_age * age
        + beta_risk * risk
    )

    prob = 1 / (1 + np.exp(-logit))
    purchase = np.random.binomial(1, prob)


    df = pd.DataFrame({
        "price": price,
        "age": age,
        "risk": risk,
        "purchase": purchase
    })
    return df
```

```python
df=new_customer_data=simulate_training_data(200)
```

```python
df.to_csv("./data/price_opt_data_1000.csv", index=False)
```

```python
def simulate_new_cases(size):
    age = np.random.uniform(20, 70, size)
    risk = np.random.uniform(0, 1, size)

    df = pd.DataFrame({
        "age": age,
        "risk": risk,
    })
    return df
```

```python
new_customer_data=simulate_new_cases(50)
```

```python
new_customer_data.to_csv("./data/new_customers_50.csv")
```

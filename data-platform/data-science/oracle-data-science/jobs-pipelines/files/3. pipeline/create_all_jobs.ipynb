# Import Packages

````python
from ads.jobs import Job, DataScienceJob, PythonRuntime
from ads import set_auth

set_auth("resource_principal")
````

# Creating the jobs

````python
job_xgboost = (
    Job(name="xgboost")
    .with_infrastructure(
        DataScienceJob()
        .with_log_group_id('<your_log_group_ocid>')
        .with_shape_name("VM.Standard.E4.Flex")
        .with_shape_config_details(memory_in_gbs=4, ocpus=1)
    )
    .with_runtime(
        PythonRuntime()
        .with_service_conda("generalml_p311_cpu_x86_64_v1")
        .with_sourc("<yout xgboost script>")
        .with_environment_variable(TREES="PLACEHOLDER",DEPTH="PLACEHOLDER",LRATE="PLACEHOLDER")
    )
)
job_xgboost.create()

job_xgboost_ocid = job_xgboost.id

job_random_forest = (
    Job(name="random_forest")
    .with_infrastructure(
        DataScienceJob()
        .with_log_group_id('<your_log_group_ocid>')
        .with_shape_name("VM.Standard.E4.Flex")
        .with_shape_config_details(memory_in_gbs=4, ocpus=1)
    )
    .with_runtime(
        PythonRuntime()
        .with_service_conda("generalml_p311_cpu_x86_64_v1")
        .with_source("<yout random forest script>")
        .with_environment_variable(TREES="PLACEHOLDER",DEPTH="PLACEHOLDER")
    )
)
job_random_forest.create()
job_random_forest_ocid = job_random_forest.id

job_comparison = (
    Job(name="comparison")
    .with_infrastructure(
        DataScienceJob()
        .with_log_group_id('<your_log_group_ocid>')
        .with_shape_name("VM.Standard.E4.Flex")
        .with_shape_config_details(memory_in_gbs=4, ocpus=1)
    )
    .with_runtime(
        PythonRuntime()
        .with_service_conda("generalml_p311_cpu_x86_64_v1")
        .with_source("<yout comparison script>")
    )
)
job_comparison.create()
job_comparison_ocid = job_comparison.id
````

````python
print('job_xgboost_ocid: ',job_xgboost_ocid) # the printed ocid are used for creating the pipeline
print('job_random_forest_ocid: ',job_random_forest_ocid)
print('job_comparison_ocid: ',job_comparison_ocid)
````

# Import Packages

````python
from ads.jobs import Job, DataScienceJob, PythonRuntime
from ads import set_auth

set_auth(auth='resource_principal') # your dynamic group should have matching rules for: data science, job, and schudeler.
````

# Creating the Job

````python
job = (
    Job(name="adult_income_model_training_basic")
    .with_infrastructure(
        DataScienceJob()
        .with_log_group_id('<your_log_group_ocid>')
        .with_shape_name("VM.Standard.E4.Flex")
        .with_shape_config_details(memory_in_gbs=4, ocpus=1)
        .with_block_storage_size(50) # minimus is 50
    )
    .with_runtime(
        PythonRuntime()
        .with_service_conda("generalml_p38_cpu_v1")
        .with_source("<yout script>") # can be a .ipynb, as well as py.
    )
)
````

````python
job.create()
````

# Running a Job Run

````python
run = job.run() # run the job
run.watch() # print the log below
````
